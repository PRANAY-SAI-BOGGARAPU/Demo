```python
from google.colab import userdata
api_key=userdata.get('GOOGLE_API_KEY_1')
```

```python
from crewai import Agent,Task,Crew,LLM
from crewai_tools import(

    FirecrawlSearchTool

)
search_tool = FirecrawlSearchTool(api_key="FIRE_CRWAL_API_KEY")
AGENT1 = Agent(
    role="Current Affairs Researcher",
    goal="Fetch top 5 current global news topics",
    backstory="You are a current affairs researcher. You are responsible for fetching the latest news topics",
    tools=[search_tool],
    llm=LLM(model="gemini/gemini-2.0-flash", api_key=api_key)
)
AGENT2 = Agent(
    role="summarizer",
    goal="summarize the dat into single line for each topic",
    backstory="You are a summarizer. You are responsible for summarizing the data into single line for each topic",
    llm=LLM(model="gemini/gemini-2.0-flash", api_key=api_key)
)
task1 =Task(
    description="Fetch top 5 current global news topics",
    expected_output="A list of top 5 current global news topics",
    agent=AGENT1
)
task2= Task(
    description="summarize the dat into single line for each topic in newline for each topic",
    expected_output="A list of top 5 current global news topics in bullet point format",
    agent=AGENT2
)
crew=Crew(
    name="Current Affairs Researcher",
    description="You are a current affairs researcher. You are responsible for fetching the latest news topics",
    agents=[AGENT1, AGENT2],
    tasks=[task1, task2],
)
print(crew.kickoff())
```
